```python
t = np.linspace(0, 2*np.pi, 100)
ysin, ycos = np.sin(t), np.cos(t)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle(None)
ax1.plot(t, ysin, color="C1")
ax2.plot(t, ycos, color="C0")
ax1.set_title("sin(t)")
ax2.set_title("cos(t)")
ax1.set_xlabel("time (s)")
ax2.set_xlabel("time (s)")
ax1.set_ylabel("amplitude (A)");
```

```python
X = np.linspace(0.1, 10*np.pi, 1000)
Y = np.sin(X)

plt.plot(X, Y, "C1o:", markevery=50);
```

```python
fig, ax = plt.subplots()
ax.set_xscale("log")

X = np.linspace(0.1, 10*np.pi, 1000)
Y = np.sin(X)

ax.plot(X, Y, "C1o-", markevery=50);
```

```python
fig, ax = plt.subplots()

X = np.linspace(0.1, 10*np.pi, 1000)
Y = np.sin(X)

ax.plot(X, Y, "C1-")
ax.fill_betweenx([-1, 1], [0], [2*np.pi], color="#ddffff")
ax.text(0, -1, r"Period $\Phi$");
```
